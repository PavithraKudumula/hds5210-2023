### 2. Query some data

Assuming the source database has some data in it or that you can load some data into it from any source, query it.  Then, read that data into a Pandas dataframe.

```python
from google.cloud import bigquery
import pandas as pd

project_id = "my-project-01-400323"
client = bigquery.Client(project=project_id)

sql_query = """
SELECT *
FROM my-project-01-400323.Pavithra_Syntheadata.encounters
"""
```

**QUERY OF DATA USING SQL**:This runs a BigQuery SQL query to select all data from the encounters table and display the result in Colab.

### 3. Aggregate your data frame

Do some kind of aggregation on your data frame.  Something that makes sense and has some groups to it.  Don't just sum up one column for the entire data frame.  Be more creative than that.

```python
df = client.query(sql_query).to_dataframe()

emergency_visits = df[df['ENCOUNTERCLASS'] == 'emergency']
```

```python
emergency_visits = emergency_visits.drop(columns=['START' , 'STOP'])
```

1. I have loaded the BigQuery data into a Pandas DataFrame for analysis
